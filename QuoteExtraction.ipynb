```python
import requests
from bs4 import BeautifulSoup as bs
import csv

x = 8

def Quotes(quote):
    with open("quotes.txt", "a+") as file:
        file.write(quote + "." + "\n")

def Quotes(quote):
    with open("quotes.txt", "a+") as file:
        file.write(quote + "." + "\n")

def convert_to_csv():
    with open("quotes.txt", "r") as file:
        quotes = file.readlines()
    quotes = [quote.strip() for quote in quotes]
    with open("allquotes.csv", "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerows([[quote] for quote in quotes])

for quotes_topic in quotes_topics:
    page = 1  # Reset page for each topic
    while x > page:
        url = f"https://www.brainyquote.com/search_results?q={quotes_topic}&pg={page}"
        r = requests.get(url)
        soup = bs(r.content, "html.parser")
        anchor1 = soup.find_all('a', {"class": "b-qt"})
        anchor2 = soup.find_all('a', {"class": "bq-aut"})
        for i, (quote_elem, author_elem) in enumerate(zip(anchor1, anchor2)):
            quote = quote_elem.find('div').text.strip()
            author = author_elem.text.strip()
            full_quote = f"{quote} - {author}"
            Quotes(full_quote)
            print(full_quote)
        page += 1

convert_to_csv()

print("Conversion to CSV and saving file completed!")
```
